# Exercise: `wc` (word count)

1. For a file (`wcfile.txt`), print out:
    - Number of lines (including blank lines)
    - Number of characters (including whitespace)
    - Number of words (separated by whitespace)
    - Number of *different* words in the file

```python
s = '_networkd:*:24:24:Network Services:/var/networkd:/usr/bin/false'

s.split(':')
```

```python
s.split(':')[0]
```

```python
s.split()  # all whitespace, any length, any combination == delimiter
```

```python
# initialize our counts
counts = {'lines':0,
          'words':0,
          'chars':0}

unique_words = set()  # start with an empty set


# loop over the file
for one_line in open('wcfile.txt'):
    counts['lines'] += 1
    counts['chars'] += len(one_line)          # len on a string, get # of characters
    counts['words'] += len(one_line.split())  # len on a list, get # of words
    unique_words.update(one_line.split())     # take words from current line, add them all to the set
    
counts['unique'] = len(unique_words)          # count elements in unique_words, and assign to our dict


# print the results
for key, value in counts.items():
    print(f'{key}: {value}')
```

# Options for `open`

- `r` -- reading (**default**)
- `w` -- writing, removing anything that was there before
- `a` -- append (write to end of file)
- `x` -- writing, but only to a new file

```python
f = open('myfile.txt', 'w')
f.write('abcd\n')
f.write('ef\n')
f.write('gh\n')
f.close()
```

```python
with open('myfile.txt', 'w') as f:
    # hidden call to f.__enter__()   ("double underscore" == "dunder")
    f.write('abcd\n')
    f.write('ef\n')
    f.write('gh\n')
    # hidden call to f.__exit__()    # flush + close
```

```python
!cat wcfile.txt
```

```python
# let's reverse each line of wcfile.txt

# with OBJECT as NAME:
# NAME = OBJECT

with open('reverse-wcfile.txt', 'w') as outfile:
    for one_line in open('wcfile.txt'):
        outfile.write(one_line.rstrip()[::-1] + '\n')   # [::-1] reverses a string
```

# List of lists

```python
movies = [['Spotlight', 129, 'Tom McCarthy'],
          ['The Big Short', 130, 'Adam McKay'],
          ['The Martian', 141, 'Ridley Scott']]

look_for = input('Movie name: ').strip()
found_it = False

for one_movie in movies:
    if look_for == one_movie[0]:  # if the names are the same...
        print(f'Found {look_for}, dir {one_movie[2]}, {one_movie[1]} minutes')
        found_it = True
        break
        
if not found_it:
    print(f'Did not find {look_for}')
```

```python
# another way: with for .. else

movies = [['Spotlight', 129, 'Tom McCarthy'],
          ['The Big Short', 130, 'Adam McKay'],
          ['The Martian', 141, 'Ridley Scott']]

look_for = input('Movie name: ').strip()

for one_movie in movies:
    if look_for == one_movie[0]:  # if the names are the same...
        print(f'Found {look_for}, dir {one_movie[2]}, {one_movie[1]} minutes')
        break
        
else:    # this is part of for..else -- runs if we reached the end of the for loop without "break"
    print(f'Did not find {look_for}')
```

# List of dicts



```python
# big improvement: use names, not numbers, to retrieve our values

movies = [{'name':'Spotlight', 'min':129, 'dir':'Tom McCarthy'},
          {'name':'The Big Short', 'min':130, 'dir':'Adam McKay'},
          {'name':'The Martian', 'min':141, 'dir':'Ridley Scott'}]

look_for = input('Movie name: ').strip()

for one_movie in movies:
    if look_for == one_movie['name']: 
        print(f'Found {look_for}, dir {one_movie["dir"]}, {one_movie["min"]} minutes')
        break
        
else:    # this is part of for..else -- runs if we reached the end of the for loop without "break"
    print(f'Did not find {look_for}')
```

# Dict of dicts

```python
# big improvement: speed of searching

movies = {'Spotlight' :  {'min':129, 'dir':'Tom McCarthy'},
          'The Big Short': {'min':130, 'dir':'Adam McKay'},
          'The Martian': {'min':141, 'dir':'Ridley Scott'}}

look_for = input('Movie name: ').strip()

if look_for in movies:
    one_movie = movies[look_for]   # inner dict, with movie info
    print(f'Found {look_for}, dir {one_movie["dir"]}, {one_movie["min"]} minutes')
        
else:
    print(f'Did not find {look_for}')
```
